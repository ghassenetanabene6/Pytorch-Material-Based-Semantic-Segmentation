```python
import PIL
import urllib
!pip install tqdm
def download_img(url,image_file_path):
    PIL.Image.open(io.BytesIO((urllib.request.urlopen(url).read()))).save(image_file_path)
```

```python
import tqdm
def scraping_imgs(image_set_path,material_class,output) :
    
    if not os.path.exists(output+'/{}'.format(material_class)):
        os.makedirs(output+'/{}'.format(material_class))
        print('create {} output directory.'.format(material_class))

    if not os.path.exists(output+'/{}/JPEG{}'.format(material_class,material_class)):
        os.makedirs(output+'/{}/JPEG{}'.format(material_class,material_class))
        print('create {} output directory.'.format(material_class,material_class)) 
        
    with open(image_set_path,'r') as img_set:
            lines=img_set.readlines()
            num=0
            k=0
            for img in lines:   
                if k%2==0:
                    try:
                        print(num)
                        download_img(img,output+'/{}/JPEG{}/{}_{}.jpg'.format(material_class,material_class,material_class,str(num).zfill(6)))
                        num+=1
                    except:
                        print('error in img : ',img)
                k+=1
                
```

```python
from shutil import copyfile
def move(path01,path02,path1,path2,path3):    
    if not os.path.exists(path1):
        os.makedirs(path1)
        print('create {} output directory.'.format(path1))
    
    if not os.path.exists(path2):
        os.makedirs(path2)
        print('create {} output directory.'.format(path2))
    if not os.path.exists(path3):
        os.makedirs(path3)
        print('create {} output directory.'.format(path3))

    file_img=open(path3+"images.txt","w")    
    res= os.walk(path01)
    i=1
    for root,dirs,files in res:
        for f in files:
            if (i<101):
                try:
                    #!mv path01/f path1/f
                    #f=f.replace('.jpg','.png')
                    #!mv path02/f path2/f
                    
                    copyfile(path01+"/"+f, path1+"/"+f)
                    f=f.replace(".jpg",".png")
                    copyfile(path02+"/"+f, path2+"/"+f)
                    f=f.replace("png","")
                    write(file_img+'\n')
                    i+=1
                except:
                    print("erreur dans "+f)
                    pass
            else:
                break
    file_img.close()
```
